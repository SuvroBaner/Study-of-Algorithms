# Here is a Fraction class implementing various operators.

```python
### Greatest Common Divisor ###
def gcd(m, n):
    while not(m % n == 0):
        oldm = m
        oldn = n
        
        m = oldn
        n = oldm % oldn
    return n

### The Fraction class starts from here ###
class Fraction:
    
    # Constructor
    def __init__(self, top, bottom):
        if (isinstance(top, int) and isinstance(bottom, int)):
            if (bottom != 0):
                common = gcd(top, bottom)
                self.num = top // common
                self.den = bottom // common
            else:
                raise RuntimeError("The denominator should not be zero")
        else:
            raise RuntimeError("Please enter integers in numerator & denominator")
    
    def __str__(self):
        if self.den == 1:
            return str(self.num)
        if self.num == 0:
            return str(0)
        return str(self.num) +'/'+ str(self.den)
    
    def __repr__(self):
        return "Num: %s, Den: %s" %(self.num, self.den)
    
    def getNum(self):
        return self.num
    
    def getDen(self):
        return self.den
    
    def __add__(self, otherfraction):
        newnum = self.num*otherfraction.den + otherfraction.num*self.den
        newden = self.den*otherfraction.den
        return Fraction(newnum, newden)
    
    def __radd__(self, otherfraction): # reverse addition
        if otherfraction == 0:
            return self
        else:
            return self.__add__(otherfraction)
        
    def __iadd__(self, otherfraction):
        self.num = self.num*otherfraction.den + self.den*otherfraction.num
        self.den = self.den*otherfraction.den
        return self
    
    def __sub__(self, otherfraction):
        newnum = self.num*otherfraction.den -  otherfraction.num*self.den
        newden = self.den*otherfraction.den
        return Fraction(newnum, newden)
    
    def __mul__(self, otherfraction):
        newnum = self.num*otherfraction.num
        newden = self.den*otherfraction.den
        return Fraction(newnum, newden)
    
    def __truediv__(self, otherfraction):
        newnum = self.num*otherfraction.den
        newden = self.den*otherfraction.num
        return Fraction(newnum, newden)
    
    def __eq__(self, other):
        firstnum = self.num*other.den
        secondnum = other.num*self.den
        return firstnum == secondnum
    
    def __ne__(self, other):
        firstnum = self.num*other.den
        secondnum = other.num*self.den
        return firstnum != secondnum
    
    def __gt__(self, other):
        firstnum = self.num*other.den
        secondnum =  other.num*self.den
        return firstnum > secondnum
    
    def __ge__(self, other):
        firstnum = self.num*other.den
        secondnum = other.num*self.den
        return firstnum >= secondnum
    
    def __lt__(self, other):
        firstnum = self.num*other.den
        secondnum =  other.num*self.den
        return firstnum < secondnum
    
    def __le__(self, other):
        firstnum = self.num*other.den
        secondnum =  other.num*self.den
        return firstnum <= secondnum
```

```python
def main():
    F1 = Fraction(1, 5)
    F2 = Fraction(3, 4)
    F3 = Fraction(1, 8)
    print("F1 = ", F1)
    print("F2 = ", F2)
    print("F3 = ", F3)
    print("Numerator of F1: ", F1.getNum())
    print("Denominator of F1: ", F1.getDen())
    print("Checking __repr__(), using a method on F2: ", F2.getDen)
    print("Is F1 >= F2 :", F1 >= F2)
    print("Is F1 <= F2 :", F1 <= F2)
    print("Is F1 == F2 :", F1 == F2)
    print("Is F1 != F2 :", F1.__ne__(F2))
    print("Addition: F1 + F2 = ", F1.__add__(F2))
    print("Adding F1, F2, F3 using sum(), it tests __radd__(): ", sum([F1, F2, F3]))
    print("Adding incrementally F1+=F2 (in variable), __iadd__(): ", F1.__iadd__(F2))
    print("F1 value has now  changed to", F1)
    print("Subtraction:  F1 - F2 = ", F1.__sub__(F2))
    print("Multiplication: F1 * F2 = ", F1.__mul__(F2))
    print("True Division: F1 / F2 = ", F1.__truediv__(F2))

if __name__ == '__main__':
    main()
```
